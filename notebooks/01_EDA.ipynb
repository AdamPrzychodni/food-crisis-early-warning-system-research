## 3. Inspecting the Geographic Data

Finally, let's load the geographic taxonomy files. These are saved as pickle files, which are used for serializing Python objects. We'll need the pickle library for this.

English and Arabic Location Names

```python
import pickle

# Load the English location names
with open('../data/id_english_location_name.pkl', 'rb') as f:
    eng_locations = pickle.load(f)

# Load the Arabic location names
with open('../data/id_arabic_location_name.pkl', 'rb') as f:
    ara_locations = pickle.load(f)

# Display a few items from each dictionary to see the structure
print("\nEnglish Locations (Sample):")
print(list(eng_locations.items())[:5])

print("\nArabic Locations (Sample):")
print(list(ara_locations.items())[:5])

print(f"\nTotal English locations: {len(eng_locations)}")
print(f"Total Arabic locations: {len(ara_locations)}")
```

## 4. Checking the Date Range

```python
# Convert the 'date' column to datetime objects for both dataframes
df_eng['date'] = pd.to_datetime(df_eng['date'])
df_ara['date'] = pd.to_datetime(df_ara['date'])

# Find the minimum and maximum dates for the English articles
eng_min_date = df_eng['date'].min()
eng_max_date = df_eng['date'].max()

# Find the minimum and maximum dates for the Arabic articles
ara_min_date = df_ara['date'].min()
ara_max_date = df_ara['date'].max()

print("English News Articles Date Range:")
print(f"From: {eng_min_date.date()} To: {eng_max_date.date()}")

print("\nArabic News Articles Date Range:")
print(f"From: {ara_min_date.date()} To: {ara_max_date.date()}")
```
