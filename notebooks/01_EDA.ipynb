## Visualizing Article Distribution by Country

Next, we'll create a bar chart to show how many articles are associated with each country in the Mashriq region. This requires us to perform a basic form of geotagging by searching for country names within the article text.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def create_location_to_country_map(locations_dict):
    location_map = {}
    for country_code, names in locations_dict.items():
        if isinstance(names, list):
            for name in names:
                location_map[name.lower()] = country_code
        else: # numpy array
             for name in names:
                location_map[name.lower()] = country_code
    return location_map

eng_location_map = create_location_to_country_map(eng_locations)
ara_location_map = create_location_to_country_map(ara_locations)

def find_country(text, location_map):
    for name, key in location_map.items():
        if name in text.lower():
            return key[:2]
    return None

df_eng['country_code'] = df_eng['body'].apply(lambda x: find_country(x, eng_location_map))
df_ara['country_code'] = df_ara['body'].apply(lambda x: find_country(x, ara_location_map))

eng_country_counts = df_eng['country_code'].value_counts()
ara_country_counts = df_ara['country_code'].value_counts()

combined_counts = pd.DataFrame({'English': eng_country_counts, 'Arabic': ara_country_counts}).fillna(0)


# --- NEW: Create a mapping from country code to full name ---
country_code_to_name = {
    'iq': 'Iraq',
    'jo': 'Jordan',
    'lb': 'Lebanon',
    'ps': 'Palestine',
    'sy': 'Syria'
}

# Apply the mapping to the DataFrame's index 
combined_counts.index = combined_counts.index.map(country_code_to_name)


# Create the bar chart using the new full-name index
combined_counts.plot(kind='bar', figsize=(12, 7), color=['skyblue', 'salmon'])
plt.title('Number of Articles per Country', fontsize=16)
plt.xlabel('Country', fontsize=12) # Changed label from 'Country Code'
plt.ylabel('Number of Articles', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Language')
plt.tight_layout()
plt.show()
```

### Risk Factor Frequency Analysis

Let's begin the advanced analysis by examining the frequency of the 167 predefined risk factors within the English news articles. This will help us identify which potential drivers of food insecurity are most prominently featured in the media coverage.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer

# Get the list of English risk factors
risk_factors = df_risk['risk_factor_english'].tolist()

# Use CountVectorizer with a predefined vocabulary
print("Counting risk factor frequencies...")
vectorizer = CountVectorizer(
    vocabulary=risk_factors,
    lowercase=True,
    ngram_range=(1, 3)
)
X = vectorizer.fit_transform(df_eng['body'].astype(str))
counts = X.sum(axis=0).tolist()[0]

df_risk_counts = pd.DataFrame({
    'Risk Factor': vectorizer.get_feature_names_out(),
    'Count': counts
}).sort_values('Count', ascending=False)
print("Counting complete.")


# Plot the top 20 most frequent risk factors
plt.figure(figsize=(12, 8))

# --- THIS IS THE CORRECTED LINE ---
sns.barplot(x='Count', y='Risk Factor', data=df_risk_counts.head(20), palette='viridis', hue='Risk Factor', legend=False)

plt.title('Top 20 Most Frequent Risk Factors in English News', fontsize=16)
plt.xlabel('Frequency Count', fontsize=12)
plt.ylabel('Risk Factor', fontsize=12)
plt.tight_layout()
plt.show()
```

### N-gram Analysis
Now, let's look for common phrases (bigrams and trigrams) in both the English and Arabic text. This can help us discover important concepts that aren't captured by the single-word risk factors.

```python
from sklearn.feature_extraction.text import CountVectorizer

# Function to get top n-grams
def get_top_ngrams(corpus, n, top_k=20):
    vec = CountVectorizer(ngram_range=(n, n), stop_words='english').fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:top_k]

# Get top bigrams for English articles
top_eng_bigrams = get_top_ngrams(df_eng['body'].astype(str), 2)
df_eng_bigrams = pd.DataFrame(top_eng_bigrams, columns=['Bigram', 'Count'])

# Get top trigrams for English articles
top_eng_trigrams = get_top_ngrams(df_eng['body'].astype(str), 3)
df_eng_trigrams = pd.DataFrame(top_eng_trigrams, columns=['Trigram', 'Count'])

# Plot the results
fig, axes = plt.subplots(1, 2, figsize=(18, 8))
sns.barplot(ax=axes[0], x='Count', y='Bigram', data=df_eng_bigrams, palette='plasma')
axes[0].set_title('Top 20 Bigrams in English News')

sns.barplot(ax=axes[1], x='Count', y='Trigram', data=df_eng_trigrams, palette='magma')
axes[1].set_title('Top 20 Trigrams in English News')

plt.tight_layout()
plt.show()
```

### Geospatial Data Density (Placeholder)

Creating a choropleth map requires a shapefile for the Mashriq region, which is not provided. The code below serves as a template for how such a visualization would be created. It would involve:
1. Loading the shapefile with `geopandas`.
2. Geotagging each article to a district.
3. Aggregating the number of articles or risk factors per district.
4. Merging this data with the GeoDataFrame.
5. Plotting the result.

```python
import geopandas as gpd
import pandas as pd
import requests
import zipfile
import io
import os
import matplotlib.pyplot as plt
import ahocorasick
import numpy as np # Import numpy for log calculation

# --- Shapefile download and loading (No changes here) ---
countries = ['IRQ', 'JOR', 'LBN', 'PSE', 'SYR']
shapefile_paths = []
data_dir = "../data/01_raw/shapefiles"
if not os.path.exists(f"{data_dir}/{countries[0]}"):
    print("Shapefiles not found, please run the download cell to get them.")
    mashriq_map = None
else:
    print("Shapefiles directory found.")
    for country in countries:
         for filename in os.listdir(f"{data_dir}/{country}"):
            if filename.endswith('_2.shp'):
                shapefile_paths.append(f"{data_dir}/{country}/{filename}")
    if not shapefile_paths:
        print("ERROR: Could not find any _2.shp files in the directories.")
        mashriq_map = None
    else:
        mashriq_map = pd.concat([gpd.read_file(path) for path in shapefile_paths], ignore_index=True)
        print("\nShapefiles loaded and combined successfully.")

# --- Aho-Corasick automaton and geotagging (No changes here) ---
if mashriq_map is not None:
    print("\nBuilding the Aho-Corasick search automaton...")
    A = ahocorasick.Automaton()
    for key, names in eng_locations.items():
        if len(key.split('_')) == 3:
            for name in names:
                A.add_word(name.lower(), key)
    A.make_automaton()
    print("Automaton built successfully.")

    def find_all_districts_fast(text):
        found_districts = set()
        for end_index, key in A.iter(text.lower()):
            found_districts.add(key)
        return list(found_districts) if found_districts else None

    print("Geotagging articles...")
    df_eng['districts'] = df_eng['body'].apply(find_all_districts_fast)
    print("Geotagging complete.")

    # --- Aggregation and Merging Logic (No changes here) ---
    df_eng_exploded = df_eng.dropna(subset=['districts']).explode('districts')
    df_geolocated = df_eng_exploded.groupby('districts').size().reset_index(name='article_count')
    key_to_name_map = {k: v[0] for k, v in eng_locations.items() if len(k.split('_')) == 3}
    df_geolocated['DISTRICT_NAME'] = df_geolocated['districts'].map(key_to_name_map)
    mashriq_map['NAME_2_lower'] = mashriq_map['NAME_2'].str.lower()
    df_geolocated['DISTRICT_NAME_lower'] = df_geolocated['DISTRICT_NAME'].str.lower()
    merged_map = mashriq_map.merge(
        df_geolocated,
        left_on='NAME_2_lower',
        right_on='DISTRICT_NAME_lower',
        how='left'
    ).fillna(0)

    # --- IMPROVEMENT 1: Plotting with a Logarithmic Scale ---
    # This helps visualize variation when outliers are present.
    # We add 1 to the count to avoid issues with log(0).
    merged_map['article_count_log'] = np.log1p(merged_map['article_count'])

    fig, ax1 = plt.subplots(1, 1, figsize=(15, 15))
    merged_map.plot(column='article_count_log', cmap='viridis', linewidth=0.8, ax=ax1, edgecolor='0.8', legend=True,
                    legend_kwds={'label': "Log of Number of News Articles", 'orientation': "horizontal"})
    ax1.set_title('News Article Density in the Mashriq Region (Log Scale)', fontdict={'fontsize': '16', 'fontweight': '3'})
    ax1.set_axis_off()
    plt.show()

    # --- IMPROVEMENT 2: Calculating and Plotting a "News Attention" Index ---
    # This normalizes the data to show which districts get a disproportionate amount of coverage.
    df_geolocated['country_code'] = df_geolocated['districts'].str[:2]
    country_totals = df_geolocated.groupby('country_code')['article_count'].transform('sum')
    df_geolocated['attention_score'] = df_geolocated['article_count'] / country_totals
    
    # Merge this new score into our map data
    attention_map = mashriq_map.merge(
        df_geolocated[['DISTRICT_NAME_lower', 'attention_score']],
        left_on='NAME_2_lower',
        right_on='DISTRICT_NAME_lower',
        how='left'
    ).fillna(0)

    fig, ax2 = plt.subplots(1, 1, figsize=(15, 15))
    attention_map.plot(column='attention_score', cmap='plasma', linewidth=0.8, ax=ax2, edgecolor='0.8', legend=True,
                       legend_kwds={'label': "News Attention Score (District Articles / Country Total)", 'orientation': "horizontal"})
    ax2.set_title('News Attention Score per District', fontdict={'fontsize': '16', 'fontweight': '3'})
    ax2.set_axis_off()
    plt.show()
```

### Risk Cluster Correlation Heatmap
Let's investigate which types of risks tend to be mentioned together in the same articles. We'll create a correlation matrix of the thematic risk clusters and visualize it as a heatmap. This will help us identify potential multicollinearity between different risk categories.

```python
# Merge the risk factors with their categories
df_risk_full = pd.merge(df_risk, df_risk_cat, left_on='risk_factor_english', right_on='risk_factor')

# Create a dictionary mapping each risk factor to its cluster
factor_to_cluster = pd.Series(df_risk_full.cluster.values, index=df_risk_full.risk_factor_english).to_dict()

# Get the unique cluster names
clusters = df_risk_full['cluster'].unique()

# Initialize a DataFrame to hold the cluster counts per article
df_cluster_counts = pd.DataFrame(0, index=df_eng.index, columns=clusters)

# This is a slow operation - we'll iterate through each cluster and count mentions
# A faster approach might use more advanced regex, but this is clear for an EDA step.
for cluster in clusters:
    factors_in_cluster = [factor for factor, c in factor_to_cluster.items() if c == cluster]
    # Create a regex pattern to find any of the factors in the cluster
    pattern = r'\b(' + '|'.join(re.escape(f) for f in factors_in_cluster) + r')\b'
    # Count the number of times any factor from the cluster appears in the body
    df_cluster_counts[cluster] = df_eng['body'].str.lower().str.count(pattern)

# Compute the correlation matrix
correlation_matrix = df_cluster_counts.corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Thematic Risk Clusters in English News', fontsize=16)
plt.show()
```

### Autocorrelation of Key Risk Themes
Finally, let's examine the temporal patterns of the most prominent risk clusters. We will plot the daily frequency of the top 5 clusters from our correlation analysis and then check for autocorrelation. This can reveal if a spike in reporting on one day is related to the reporting on previous days.

```python
from pandas.plotting import autocorrelation_plot

# Sum the cluster counts to find the most frequent clusters
top_5_clusters = df_cluster_counts.sum().sort_values(ascending=False).head(5).index.tolist()

# Create a new DataFrame with daily counts for the top 5 clusters
df_daily_clusters = df_eng[['date']].copy()
df_daily_clusters = pd.concat([df_daily_clusters, df_cluster_counts[top_5_clusters]], axis=1)
df_daily_clusters = df_daily_clusters.groupby('date').sum()

# Plot the time series of the top 5 clusters
df_daily_clusters.plot(figsize=(15, 7), subplots=False,
                       title='Daily Mentions of Top 5 Risk Clusters')
plt.ylabel('Number of Mentions')
plt.show()

# Plot the autocorrelation for each of the top 5 clusters
fig, axes = plt.subplots(len(top_5_clusters), 1, figsize=(10, 15), sharex=True)
for i, cluster in enumerate(top_5_clusters):
    autocorrelation_plot(df_daily_clusters[cluster], ax=axes[i])
    axes[i].set_title(f'Autocorrelation for: {cluster}')

plt.tight_layout()
plt.show()
```
